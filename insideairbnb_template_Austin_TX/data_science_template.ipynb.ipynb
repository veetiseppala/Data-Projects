#### Siivoamisen jälkeen tehdään malli siivotun datan avulla
Tässä esimerkissä käytämme yksinkertaista lineaarista regressiomallia ennustamaan hintaa (price) muiden muuttujien perusteella. Tämä on perusmalli, jota voidaan käyttää hintojen ennustamiseen perustuen eri ominaisuuksiin kuten sijaintiin  Malli on hieman samanlainen kuin TampereBnB priceprediction tehtävässä


```python
categorical_features = ['room_type']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")

# Valitaan mallin muuttujat
features = data[['latitude', 'longitude', 'room_type', 'availability_365', 'neighbourhood']]
target = data['price']

# Jako opetus- ja testidataan
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Luodaan ja koulutetaan lineaarinen regressiomalli
model = Pipeline(steps=[('preprocessor', transformer),
                        ('model', LinearRegression())])

model.fit(X_train, y_train)

# Ennustetaan testidatalla
y_pred = model.predict(X_test)

# Lasketaan mallin suorituskyky
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R² Score: {r2}")
```

Näiden tulosten perusteella voimme päätellä, että malli ei suoriudu hyvin hintojen ennustamisessa. Tämä voi johtua monista tekijöistä, kuten riittämättömästä tai epärelevantista ominaisuusvalikoimasta, yksinkertaisesta mallityypistä (lineaarinen regressio), joka ei sovi monimutkaisiin ei-lineaarisiin suhteisiin datassa, tai liian vähäisestä datan käsittelystä ennen mallin koulutusta.

```python
# Kokeillaan uudestaan hyödyntämällä Price prediction tehtävässä käyttämääni mallia

# Yhden kuuman koodaus huonetyypille
categorical_features = ['room_type']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)], remainder="passthrough")

# Valitaan mallin muuttujat
features = data[['latitude', 'longitude', 'room_type', 'availability_365']]
target = data['price']

# Jako opetus- ja testidataan
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Luodaan ja koulutetaan RandomForestRegressor-malli
model = Pipeline(steps=[('preprocessor', transformer),
                        ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))])

model.fit(X_train, y_train)

# Ennustetaan testidatalla
y_pred = model.predict(X_test)

# Lasketaan mallin suorituskyky
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R² Score: {r2}")
```

Vaikka RandomForestRegressor ei suoriutunut virheettömästi (kuten RMSE ja MSE osoittavat), se on selkeästi parempi malli tässä tapauksessa, koska se selittää hintavarianssin huomattavasti paremmin kuin lineaarinen regressiomalli. Tämä osoittaa, että monimutkaisemmat mallit voivat olla hyödyllisiä monimutkaisempien suhteiden ja piirteiden mallintamisessa datassasi.

### Tehdään scraperillä tutkimusta Austinin zip koodien ja keskimäärästen ansioiden välillä

```python
# URL sivulle
url = "http://zipatlas.com/us/tx/austin/zip-code-comparison/highest-median-household-income.htm"

# HTTP GET request
response = requests.get(url)

# BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Etsitään taulu id peruusteella
table = soup.find('table', {'id': 'comp'})
```

```python
# Tallenna ZIP-koodit ja taloudelliset tulot listaan
zip_income_list = []

# Käy läpi jokainen taulun rivi
for row in table.find_all('tr')[1:]: # Hypätään ensimmäinen rivi, koska se sisältää otsikot
    cells = row.find_all('td')
    if len(cells) > 1:  
        zip_code = cells[1].text.strip()
        income = cells[2].text.strip()
        zip_income_list.append((zip_code, income))

# Tulostetaan ZIP-koodit ja taloudelliset tulot
for item in zip_income_list:
    print(f"ZIP Code: {item[0]}, Household Income: {item[1]}")
```

```python
# Yhdistään aiempi data
zip_income_data = pd.DataFrame(zip_income_list, columns=['ZIP Code', 'Median Household Income'])

# Poista '$'-merkit ja muunna tulot kokonaisluvuiksi
zip_income_data['Median Household Income'] = zip_income_data['Median Household Income'].replace('[\$,]', '', regex=True).astype(int)
zip_income_data['ZIP Code'] = zip_income_data['ZIP Code'].astype(int)

# Yhdistä data käyttäen 'neighbourhood'-saraketta yhdistävänä avaimena
merged_data = pd.merge(data, zip_income_data, left_on='neighbourhood', right_on='ZIP Code', how='left')

print(merged_data[['neighbourhood', 'Median Household Income']].head())
```

```python
# Tutkitaan onko airbnb hinnan ja taloudellisen tulotason välillä korrelaatiota

X_train, X_test, y_train, y_test = train_test_split(merged_data[['Median Household Income']], merged_data['price'], test_size=0.2, random_state=0)

# Luo lineaarinen regressiomalli
model = LinearRegression()
model.fit(X_train, y_train)

# Tee ennusteet
predictions = model.predict(X_test)

# Laske korrelaatiokerroin
r2 = r2_score(y_test, predictions)

print(f'Korrelaatiokerroin (R^2): {r2}')
```

Huomataan ettei alueella ja airbnb hinnalla ole suurta korrelaatiota. Tämän lisäksi voitaisiin luoda vielä tarkempia malleja, kuten aikaisemmin ja sen avulla saada parempaa tietoa siitä, mitkä vaikuttavat airbnb asuntojen hintaan.
